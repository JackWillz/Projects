```python
import os
os.chdir('C:/Users/Jack Wills/Documents/marvel')

from bs4 import BeautifulSoup
import requests
import re

# Base soupify function to return content
def soupify(link):
    r = requests.get(link)
    soup = BeautifulSoup(r.content)
    return soup

# Finds the article within the soup of the characters page
def extract_article(soup):
    content = soup.find(id='mw-content-text').findAll('p')
    ignores = ['<a', '<span', 'caption', '\n']
    article_text = ''
    for element in content:
        if any(i in str(element)[:18] for i in ignores) or ':' in str(element)[:20] or 'History of character is unknown' in str(element):
            continue
        else:
            text = element.findAll(text = True)
            article_text += ''.join(text)
    article_text = re.sub(r"[\(\[].*?[\)\]]", "", article_text)
    article_text = article_text.replace('\n', ' ')
    article_text = article_text.replace('  ', ' ')
    return article_text

# Create all the characters on a single page
def find_characters(letter):
    link = 'https://marvel.fandom.com/wiki/Category:Earth-616_Characters?from=' + letter
    soup = soupify(link)
    characters = []
    content = soup.findAll('a', {'class': 'category-page__member-link'})
    for element in content:
        if '/wiki/' in str(element):
            result = re.search('href="(.*)" ', str(element))
            characters.append(result.group(1))
        else:
            continue
    return characters

# Iterate through all pages to find all the characters
def find_all_characters()
    start = 'AAA'
    all_characters = []
    repeat = False
    check_repeat = 'None'
    all_characters.extend(find_characters(start))
    while repeat is False:
        last_character = all_characters[-1:][0].replace('/wiki/','')
        print(last_character)
        if last_character == check_repeat:
            repeat == True
            break
        else:
            check_repeat = last_character
            new_characters = find_characters(last_character)[1:]
            all_characters.extend(new_characters)
    return all_characters

# For every character, find the bio
def get_all_articles(all_characters):
    all_articles = ''
    for character in all_characters:
        print(character)
        character_link = 'https://marvel.fandom.com' + character
        soup = soupify(character_link)
        article = extract_article(soup)
        all_articles += ''.join(article)
    
    return all_articles

```

```python
all_articles = get_all_articles(all_characters)
```

```python
with open('marvel_data.txt', 'w', encoding='utf-8') as file:
    file.write(all_articles)
```
